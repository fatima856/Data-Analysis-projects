```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv('Processsed_Superstore_Sales_Data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df1=df.dropna()
df1.isna().sum()
```

```python
df1.duplicated().sum()
```

```python
df1.to_csv('Processsed_Superstore_Sales_Data.csv',index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
#What is the top 6 cities in Sales ??
city_sales = df1.groupby('City')['Sales'].sum().sort_values(ascending=False).head(6)

plt.figure(figsize=(10, 6))
city_sales.plot(kind='bar')
plt.title('Top 6 Cities by Total Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# what is the relation between state and sales ??
state_sales = df1.groupby('State')['Sales'].sum().sort_values()

plt.figure(figsize=(12, 8))
state_sales.plot(kind='bar')
plt.title('Total Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# what is the relation between region and sales ??
region_sales = df1.groupby('Region')['Sales'].sum()

plt.figure(figsize=(8, 8))
region_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Total Sales by Region')
plt.ylabel('')  # Hide the y-label
plt.tight_layout()
plt.show()
```

```python
# what is the relation between category and sales ??
category_sales = df1.groupby('Category')['Sales'].sum()

plt.figure(figsize=(8, 8))
category_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140, wedgeprops={'width': 0.3})
plt.title('Total Sales by Category')
plt.ylabel('')  # Hide the y-label
plt.tight_layout()
plt.show()
```

```python
#what is the relation between segment and sales ??
segment_sales = df1.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(8, 6))
segment_sales.plot(kind='barh')
plt.title('Total Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#what is the relation between ship mode and sales ??
ship_mode_sales = df1.groupby('Ship Mode')['Sales'].sum()

plt.figure(figsize=(8, 6))
ship_mode_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Total Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#what is the top 20 saled products ?
product_sales = df1.groupby('Product Name')['Sales'].sum().sort_values(ascending=True).tail(20)  # Top 20 products

plt.figure(figsize=(10, 8))
product_sales.plot(kind='barh')
plt.title('Top 20 Products by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```
